```python
#Task2: The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values. Replace the missing values by running average. You should use the “For” loop for this task.

import pandas as pd
import numpy as np

# Load the data (assuming the file contains the necessary columns)
df = pd.read_csv(r"C:\Users\Darshan Dinesh\Downloads\Skillo Villa\Python capestone\Project DataFrame.csv")

# Ensure the 'Cost' column is numeric and contains NaN for missing values
df['Cost'] = pd.to_numeric(df['Cost'], errors='coerce')

# Step 1: Initialize a variable to store the running total and count
running_total = 0
count = 0

# Step 2: Replace missing values with the running average using a 'for' loop
for i in range(len(df)):
    if not np.isnan(df.loc[i, 'Cost']):  # If the value is not missing
        running_total += df.loc[i, 'Cost']
        count += 1
    else:  # If the value is missing, replace it with the running average
        if count > 0:  # Avoid division by zero
            running_average = running_total / count
            df.loc[i, 'Cost'] = running_average
        else:
            df.loc[i, 'Cost'] = 0  # Default to 0 if no previous values exist

# Print the updated DataFrame
df['Cost'] = df['Cost'].apply(lambda x: f"{x:.2f}")
print(df)

```

```python
#Task 5 Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects..


import pandas as pd

# Load the Final.csv file
df = pd.read_csv(r"C:\Users\Darshan Dinesh\Downloads\Skillo Villa\Python capestone\Final.csv")

# Ensure that the 'Cost' column is numeric (handle missing or invalid values if any)
df['Cost'] = pd.to_numeric(df['Cost'], errors='coerce')

# Convert the 'Status' column to strings and handle missing values
df['Status'] = df['Status'].fillna('').astype(str).str.strip().str.lower()

# Add the 'Bonus' column with 5% of 'Cost' for Finished projects
df['Bonus'] = df.apply(
    lambda row: row['Cost'] * 0.05 if row['Status'] == 'finished' else 0,
    axis=1
)

print(df)
```

```python
# task 6 Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.

import pandas as pd

# Load the data (assuming the file contains the necessary employee and project information)
df = pd.read_csv(r"C:\Users\Darshan Dinesh\Downloads\Skillo Villa\Python capestone\Final.csv")

# Step 1: Demote the designation level by 1 for employees with 'fail' status in the 'Status' column
# Assuming 'Status' column contains the project status and 'Designation_Level' column represents the employee's level
df.loc[df['Status'] == 'fail', 'Designation Level'] -= 1

# Step 2: Delete employees whose designation level is above 4
df = df[df['Designation Level'] <= 4]

# Print the resulting DataFrame
print(df)
```

```python
#Task 7 Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

import pandas as pd

# Load the data (assuming the file contains the necessary columns)
df = pd.read_csv(r"C:\Users\Darshan Dinesh\Downloads\Skillo Villa\Python capestone\Final.csv")

# Step 1: Split the 'Name' column into 'First_Name' and 'Last_Name'
df[['First_Name', 'Last_Name']] = df['Name'].str.split(' ', n=1, expand=True)

# Step 2: Add "Mr." or "Mrs." to the first name based on gender
df['First_Name'] = df.apply(
    lambda row: f"Mr. {row['First_Name']}" if row['Gender'].lower() == 'male' else f"Mrs. {row['First_Name']}",
    axis=1
)

# Step 4: Drop the 'Gender' column
df = df.drop(columns=['Gender'])

# Print the resulting DataFrame
print(df)

```

```python
# Task 8 Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

import pandas as pd

# Load the data (assuming the file contains the necessary columns)
df = pd.read_csv(r"C:\Users\Darshan Dinesh\Downloads\Skillo Villa\Python capestone\Final.csv")

# Step 1: Promote the designation level by 1 for employees whose age is more than 29 years
df['Designation Level'] = df.apply(
    lambda row: row['Designation Level'] + 1 if row['Age'] > 29 else row['Designation Level'],
    axis=1
)

# Print the resulting DataFrame
print(df)
```

```python
#Task 9 Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost

import pandas as pd

# Load the data (assuming the file contains the necessary columns)
df = pd.read_csv(r"C:\Users\Darshan Dinesh\Downloads\Skillo Villa\Python capestone\Final.csv")

# Step 1: Ensure that the 'Cost' column is numeric (handle missing or invalid values if any)
df['Cost'] = pd.to_numeric(df['Cost'], errors='coerce')

# Step 2: Group by 'Employee_ID' (or a similar column) to sum the 'Cost' of all projects for each employee
total_proj_cost = df.groupby(['ID', 'First_Name'], as_index=False)['Cost'].sum()

# Step 3: Rename the column to 'Total Cost' for clarity
total_proj_cost = total_proj_cost.rename(columns={'Cost': 'Total Cost'})

# Step 4: Save the resulting DataFrame with ID, First Name, and Total Cost
print(total_proj_cost)
```
